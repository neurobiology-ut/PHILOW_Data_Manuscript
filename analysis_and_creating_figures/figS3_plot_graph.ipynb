```r
library(ggplot2)
```

```r
options(repr.plot.width=20, repr.plot.height=8)
```

```r
no_data <- data.frame(
    time = c(0, 1200),
    accuracy = c(0, 33.264)
)

DL_data <- data.frame(
    time = c(0, 54.25, 357.62, 556.03, 768.13),
    accuracy = c(0, 0, 98.304, 147.456, 200)
)

DL_HITL_data <- data.frame(
    time = c(0, 37.78, 149.26, 208.37, 271.55),
    accuracy = c(0, 0, 98.304, 147.456, 200)    
)

DL_3D_AL_data <- data.frame(
    time = c(0, 34.00, 55.33, 68.32, 83.2),
    accuracy = c(0, 0, 98.304, 147.456, 200)
)
```

```r
yn_DL_data <- data.frame(
    time = c(0, 1073.780696-1007.06403, 1073.780696),
    accuracy = c(0, 0, 98.304)
)

yn_DL_HITL_data <- data.frame(
    time = c(0, 544.6003052-521.7369718, 544.6003052),
    accuracy = c(0, 0, 98.304)    
)

yn_DL_3D_AL_data <- data.frame(
    time = c(0, 113.33-91.6, 113.33),
    accuracy = c(0, 0, 98.304)
)
```

```r
g <- ggplot(NULL) + theme(
    axis.line = element_line(size = 1, colour="black"),
    axis.text.x = element_text(size=16, face = "bold", family = "Helvetica", color="black"),
    axis.text.y = element_text(size=16, face = "bold", family = "Helvetica", color="black"),
    axis.title.x = element_text(size = 24, face = "bold", family = "Helvetica", color="black"),
    axis.title.y = element_text(size = 24, face = "bold", family = "Helvetica", color="black"),   
    panel.background = element_blank(), 
    panel.grid.major = element_blank(),
    panel.grid.minor = element_blank(),
) + scale_y_continuous(
    expand = c(0,0), 
    limits=c(0, 110)
) + scale_x_continuous(
    expand = c(0,0), 
    limits=c(0, 1220)
) 


# manual

g <- g + geom_line(
    data = no_data,
    aes(x = time, y = accuracy),
    size=3,
    color="gray",
) 



## Nakanishi

# DL_baseline

g <- g + geom_line(
    data = yn_DL_data[1:2,],
    aes(x = time, y = accuracy),
    size=3,
    color = "royalblue2"
) + geom_line(
    data = yn_DL_data[2:3,],
    aes(x = time, y = accuracy),
    size=3,
    color = "royalblue2"
) + geom_point(
    data = yn_DL_data[1:4,],
    aes(x = time, y = accuracy),
    size=2,
    color = "royalblue2"
) 


# HITL+

g <- g + geom_line(
    data = yn_DL_HITL_data[1:2,],
    aes(x = time, y = accuracy),
    size=3,
    color="limegreen"
) + geom_line(
    data = yn_DL_HITL_data[2:3,],
    aes(x = time, y = accuracy),
    size=3,
    color="limegreen"
) + geom_point(
    data = yn_DL_HITL_data[1:4,],
    aes(x = time, y = accuracy),
    size=2,
    color="limegreen"
)


# HITL+, TAP+

g <- g + geom_line(
    data = yn_DL_3D_AL_data[1:2,],
    aes(x = time, y = accuracy),
    size=3,
    color="deeppink2"
) + geom_line(
    data = yn_DL_3D_AL_data[2:3,],
    aes(x = time, y = accuracy),
    size=3,
    color="deeppink2"
) + geom_point(
    data = yn_DL_3D_AL_data[1:4,],
    aes(x = time, y = accuracy),
    size=2,
    color="deeppink2"
)



# make label

g <- g + labs(
    x = "Human Work Time (min)",
    y = "Segmented Volume\n(M voxel)"
) 


# annotate yokosenn

g <- g + annotate(
    "segment",
    x = 0, xend = 1200,
    y = 98.304, yend = 98.304,
    size = 0.5,
    color="grey20",
    linetype="dashed"
) 

g
```
